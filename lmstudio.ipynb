```python
import lmstudio as lms
from lint_ii import ReadabilityAnalysis
llm = lms.llm() # Get any loaded LLM
```

```python
llm
```

```python
anal = ReadabilityAnalysis.from_text("""De Oudegracht is het sfeervolle hart van de stad.
In de middeleeuwen was het hier een drukte van belang met de aan- en afvoer van goederen. 
Nu is het een prachtige plek om te winkelen en te lunchen of te dineren in de oude stadskastelen.""")

print(anal.lint.score)
print(anal.lint_scores_per_sentence)
```

```python
prompt = """
SYSTEMPROMPT:
Je bent een expert op het schrijven van teksten in vereenvoudige taal, dit zijn je instructies:
Voeg geen tekst toe. Denk eerst na, leg dan je stappen van herschrijven uit. Herschrijf alleen zin 2. Pas voor de rest niks aan. Gebruik eenvoudige taal

USER PROMPT:
De Oudegracht is het sfeervolle hart van de stad.
In de middeleeuwen was het hier een drukte van belang met de aan- en afvoer van goederen. 
Nu is het een prachtige plek om te winkelen en te lunchen of te dineren in de oude stadskastelen.

"""

prediction = llm.respond(prompt)
print(prediction.content)
```

```python
analysis = ReadabilityAnalysis.from_text(prediction.content)
print(analysis.lint.score)
print(analysis.lint_scores_per_sentence)
```
