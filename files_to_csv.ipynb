# Merge contents of .txt files into one .csv files which can be read by Pandas.
Returns prep_texts.csv

```python
import sys
import pandas as pd
from pathlib import Path
```

```python
p = Path(r"../papers/LIN_teksten_2016-6-16/")
```

```python

prep_txts = []
for f in p.glob('*.txt'):
    print(f.name[:-4])
    print(f.read_text(encoding='utf-8').strip())
    file_data = [f.name[:-4], f.read_text(encoding='utf-8').strip()]
    prep_txts.append(file_data)

```

```python
df = pd.DataFrame(prep_txts, columns=["id", "text"])
```

```python
df.head()
```

```python
df.to_csv("./prep_texts.csv", index=False)
```
